```python
#Making of pairs
import itertools
def make_pairs(X,y):
  X_pairs,y_pairs=[],[]
  tuples = [(x1, y1) for x1, y1 in zip(X, y)]
  for t in itertools.product(tuples, tuples):
    pair_A, pair_B = t
    img_A,label_A=pair_A
    img_B,label_B=pair_B

    new_label=int(label_A==label_B)
    X_pairs.append([img_A,img_B])
    y_pairs.append(new_label)
  X_pairs=np.array(X_pairs)
  y_pairs=np.array(y_pairs)
  return X_pairs,y_pairs
```

```python
X_train_pairs,y_train_pairs=make_pairs(x_train_sample,y_train_sample)
```

```python
#Build CNN model
from tensorflow.keras.layers import * #Provides different types of layers for building neural networks.
from tensorflow.keras.models import Sequential,Model #Used for stacking layers sequentially.,Used for defining custom models.
def get_cnn_bock(depth):
  return Sequential([Conv2D(depth,3,1),BatchNormalization(),ReLU()])
DEPTH=64
cnn=Sequential([Reshape((28,28,1)),get_cnn_bock(DEPTH),GlobalAveragePooling2D(),Dense(64,activation='relu')])
```

```python
imag_A=Input(shape=(28,28),name='imag_A')
imag_B=Input(shape=(28,28),name='imag_B')
```

```python
feature_A=cnn(imag_A)
feature_B=cnn(imag_B)
```

```python
#difference_layer=Lambda(lambda x:tf.math.abs(x[0]-x[1]))([feature_A,feature_B])
difference_layer=tf.keras.layers.Subtract()([feature_A,feature_B])
```

```python
dense=Dense(64,activation='relu')(difference_layer)
output=Dense(1,activation='sigmoid')(dense)
```

```python
model=Model(inputs=[imag_A,imag_B],outputs=output)
```

```python
model.summary()
```

```python
def contrastive_loss(y_true,y_pred,margin=1.0):
  positive_loss=y_true*tf.square(y_pred)
  negative_loss=(1-y_true)*tf.square(tf.maximum(margin-y_pred,0))
  return tf.reduce_mean(positive_loss+negative_loss)
```

```python
model.compile(loss=contrastive_loss,optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stopping=EarlyStopping(patience=3)
```

```python
from tensorflow.keras import backend as K

def euclidian(x,y):

  sum_square = K.sum(K.square(x - y), axis=1, keepdims=True)
  return K.sqrt(K.maximum(sum_square, K.epsilon()))
```

```python
model.compile(loss=euclidian,optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),metrics=['accuracy'])
```
